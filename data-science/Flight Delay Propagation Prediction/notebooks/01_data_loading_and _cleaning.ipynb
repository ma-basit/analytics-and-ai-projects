# Data Loading and Cleaning

Objective: Load the raw airline operations data and perform essential cleaning so that only valid, comparable flight records are passed to the next stage of analysis.

```python
import pandas as pd
```

```python
df = pd.read_parquet("../data/raw/flight_data.parquet")
df.head(5)
```

```python
df.info()
```

```python
print(f"Number of columns: {df.shape[1]}")
df.columns.tolist()
```

```python
# Remove unnamed artifact column
if "Unnamed: 119" in df.columns:
    df = df.drop(columns=["Unnamed: 119"])

# Strip whitespace from column names
df.columns = df.columns.str.strip()

df.shape[1]
```

```python
df = df[df["Cancelled"] == 0]
```

```python
df = df[df["Diverted"] == 0]
```

```python
df = df[df["Duplicate"] != "Y"]
df = df.reset_index(drop=True)

df.shape
```

```python
# Ensure no cancelled or diverted flights remain
assert df["Cancelled"].sum() == 0
assert df["Diverted"].sum() == 0

# Ensure duplicate flag is clean
assert not (df["Duplicate"] == "Y").any()

print("Integrity checks passed.")
```

```python
df.info()
```

```python
df.to_parquet("../data/processed/flight_data_clean.parquet", index=False)
```

## Summary

In this notebook, the raw flight dataset was cleaned to create a reliable base table for further analysis.

The data was loaded from a Parquet file, basic structural issues were fixed, and invalid flight records were removed. Cancelled flights, diverted flights, and duplicate operational records were filtered out. Column names were cleaned, and basic integrity checks were applied to ensure consistency.

No exploratory analysis, feature engineering, or modeling decisions were made at this stage. The resulting dataset is now ready for EDA, target definition, and feature engineering in the next notebook.

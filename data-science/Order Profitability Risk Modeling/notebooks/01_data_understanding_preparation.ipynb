```python
# Create order-level dataset

orders_df = df.groupby('Order Id').agg({
    'Sales': 'sum',
    'Order Item Total': 'sum',
    'Order Item Discount': 'sum',
    'Order Item Quantity': 'sum',
    'Order Profit Per Order': 'sum',

    'Customer Segment': 'first',
    'Market': 'first',
    'Shipping Mode': 'first',
    'Order Region': 'first',
    'Order Country': 'first'
}).reset_index()
```

```python
df.groupby('Order Id')['Order Profit Per Order'].nunique().value_counts()
```

```python
orders_df = orders_df.rename(columns={
    'Sales': 'total_sales',
    'Order Item Total': 'total_order_value',
    'Order Item Discount': 'total_discount',
    'Order Item Quantity': 'total_quantity',
    'Order Profit Per Order': 'order_profit'
})
```

## Notebook Summary & Key Learnings

### What I did in this notebook

In this notebook, I started by understanding the dataset and preparing it for further analysis.

- Loaded and explored the dataset
- Checked data types, missing values, and structure
- Realized that the dataset is at **order-item level** (multiple rows per order)
- Since my problem is about **order-level profit**, I aggregated the data to order level
- Verified how profit is stored before aggregation (important step)
- Renamed columns to make them easier to understand
- Created a clean dataset for further analysis

---

### Important thing I learned

One important thing I learned here is that I should not blindly trust column names.

Even though the column was named *"Order Profit Per Order"*, it was actually different for different items in the same order.

So I checked it using `.nunique()` and found that:
- Profit was different across items in many orders
- This means profit was stored at **item level**

Because of this, I used **sum aggregation** to calculate total order profit.

If I had used `first`, the profit calculation would have been wrong.

---

### Key Observations from Data

- The dataset contains around **65,000 orders**, which is good for analysis
- Each order has around **2–3 items on average**
- Most orders are profitable, but some have **very high losses**
- The minimum profit is around **-4300**, which is very large
- Profit values are not evenly distributed (some extreme values exist)

### Business Understanding

From a business point of view:

- Just looking at **sales is not enough**, because some orders still make losses
- Some orders are likely losing money due to:
  - high discounts
  - product combinations
  - or operational factors (like shipping)

This is why predicting **profit risk** is important.

---

### Why this step was important

This step was important because:

- The original dataset structure did not match the business problem
- If I had directly used the raw data, the model would not be meaningful
- Correct aggregation ensures that each row represents one decision unit (one order)

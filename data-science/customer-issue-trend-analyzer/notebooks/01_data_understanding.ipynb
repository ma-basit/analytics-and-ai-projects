### 01 — Dataset Understanding

#### Objective
Understand the structure of the Consumer Complaint dataset, identify key columns for analysis, and assess data quality before text preprocessing.


```python
import pandas as pd
```

```python
data_path = "../data/raw/consumer_complaints.csv"
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df[[
    "Consumer complaint narrative",
    "Date received",
]].isna().sum()
```

```python
df_non_empty = df.dropna(subset=["Consumer complaint narrative"])
```

```python
df_non_empty.shape
```

```python
df_non_empty["Consumer complaint narrative"].duplicated().sum()
```

```python
dup_examples = (
    df_non_empty[df_non_empty["Consumer complaint narrative"]
    .duplicated(keep=False)]
    .sort_values("Consumer complaint narrative")
    .head(3)
)

dup_examples[["Consumer complaint narrative", "Complaint ID"]]
```

```python
df_non_empty = df_non_empty.drop_duplicates(
    subset=["Consumer complaint narrative"]
)
df_non_empty.shape
```

```python
df_non_empty.info()
```

```python
df_non_empty["Date received"] = pd.to_datetime(df_non_empty["Date received"])
```

```python
df_non_empty.info()
```

```python
columns_to_keep = [
    "Date received",
    "Consumer complaint narrative",
    "Product",
    "Issue",
    "Sub-issue",
    "Company"
]

df_step1 = df_non_empty[columns_to_keep].copy()
```

```python
df_step1.reset_index(drop=True, inplace=True)
```

```python
output_path = "../data/processed/complaints_step1_filtered.csv"
df_step1.to_csv(output_path, index=False)
```

### Summary

- Loaded the Consumer Complaint dataset containing more than 1.2 million records.
- Identified the complaint narrative and complaint received date as the primary fields required for NLP and trend analysis.
- Examined missing values and restricted the dataset to complaints with available narrative text.
- Checked for exact duplicate complaint narratives and removed them to avoid over-representation of repeated complaints.
- Converted the complaint received date into datetime format to enable time-based analysis.
- Selected a subset of relevant columns to reduce dataset size and improve processing efficiency.
- Saved a cleaned and filtered dataset (approximately 367k complaints) for use in subsequent NLP steps.
